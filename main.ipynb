#### Circuits and moments

```python
a = cirq.NamedQubit("a")
```

```python
circuit = cirq.Circuit(
    cirq.H(a),
    cirq.S(a),
    cirq.T(a),
    cirq.X(a),
    cirq.measure(a)
)
```

```python
print(circuit)
```

```python
simulator = cirq.Simulator()
```

```python
result = simulator.run(circuit, repetitions=100)
```

```python
cirq.plot_state_histogram(result, plt.subplot())
plt.show()
```

```python
result = simulator.run(circuit, repetitions=1000)
cirq.plot_state_histogram(result, plt.subplot())
plt.show()
```
